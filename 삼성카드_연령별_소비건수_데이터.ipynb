```python
import os
import pandas as pd

path = './data/'
fname = os.listdir(path)
print(fname)
```

```python
df = pd.read_csv('./data/배달음식_선호도_2.csv', encoding='cp949')
df
```

```python
df_20_30 = df[(df['연령대']>='20대') & (df['연령대']<='30대')]
df_20_30.iloc[:,[1,3,4,5,7,9]].head(15)
```

```python
df_40_60 = df[(df['연령대']>='40대') & (df['연령대']<='60대이상')]
df_40_60.iloc[:,[1,3,4,5,7,9]].head(15)
```

```python
print('소비건수합계 :', df_40_60['소비건수합계'].sum())
print('건강관심도 등급 :', round(df_40_60['건강관심도등급'].mean()))
print('배달식선호 등급 :', round(df_40_60['배달식선호등급'].mean()))
```

```python
print('소비건수합계 :', df_20_30['소비건수합계'].sum())
print('건강관심도 등급 :', round(df_20_30['건강관심도등급'].mean()))
print('배달식선호 등급 :', round(df_20_30['배달식선호등급'].mean()))
```

```python
age_con = {'연령대':['20~30대', '40~60대이상'],
           '총_소비건수':[df_20_30['소비건수합계'].sum(), df_40_60['소비건수합계'].sum()],
           '건강관심도':[round(df_20_30['건강관심도등급'].mean()), round(df_40_60['건강관심도등급'].mean())],
           '배달식선호':[round(df_20_30['배달식선호등급'].mean()), round(df_40_60['배달식선호등급'].mean())]}
age_con
```

```python
소비건수 = pd.DataFrame(age_con)
소비건수
```

```python
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic")
#plt.rc("font", family="AppleGothic")
plt.rc("axes", unicode_minus=False)  #-폰트 깨지는 문제 해결



plt.figure(figsize=(15,8))

plt.subplot(1, 3, 1)
ax = sns.barplot(x='연령대',y='총_소비건수', data=소비건수)
ax.set_xlabel('연령', size = 12)  # x축 이름 설정
ax.set_ylabel('소비건수 합', size = 12) # y축 이름 설정
ax.set_title('연령별 소비건수', size = 15)
ax



plt.subplot(1, 3, 2)
ax1 = sns.barplot(x='연령대',y='건강관심도',data=소비건수)
ax1.set_xlabel('연령', size = 12)  # x축 이름 설정
ax1.set_ylabel('건강관심도 평균', size = 12) # y축 이름 설정
ax1.set_title('연령별 건강관심도', size = 15)
ax1



plt.subplot(1, 3, 3)
ax2 = sns.barplot(x='연령대',y='배달식선호',data=소비건수)
ax2.set_xlabel('연령', size = 12)  # x축 이름 설정
ax2.set_ylabel('배달식선호 평균', size = 12) # y축 이름 설정
ax2.set_title('연령별 배달식선호', size = 15)
ax2
```
