```python
import requests
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
import sys
import os
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time



sch_text = input('검색어 입력 : ')

driver=webdriver.Chrome(ChromeDriverManager().install())
time.sleep(2)

url = 'https://www.yna.co.kr/search/index?query=' + sch_text + '&ctype=A&sort=weight&from=20200526&to=20210526&period=1y'
driver.get(url)
time.sleep(5)


html=driver.page_source
soup=bs(html, 'html.parser')

li = soup.find('div', class_='cts_atclst').find_all('li')

textlist = []
for temp in range(1, len(li)+1):

    driver.find_element_by_xpath('//*[@id="article_list"]/div[2]/ul/li['+str(temp)+']/a/span[1]').click()
    time.sleep(2)

    html=driver.page_source
    soup=bs(html, 'html.parser')
    test = soup.find('div', class_='content01 scroll-article-zone01').find('article', class_='story-news article').text
    
    textlist.append(test)
    
    time.sleep(3)
    driver.back()   ### 뒤로가기
    time.sleep(2)

print(textlist)




driver.close()
```

```python
re_retextlist = []
for i in textlist:
    retextlist = i.replace('\n', '').replace('[농식품부제공.재판매및DB금지]', '').replace('[촬영조정호]', '').replace('연합뉴스', '').replace("'\'", '').replace('[DB 및 재판매 금지]', '').replace('광고', '').replace('특파원', '').replace('기자', '').replace('위해', '').replace('제공', '').replace('지난', '')[12:-59]
    re_retextlist.append(retextlist)

#print(re_retextlist)

df['description'] = pd.DataFrame(re_retextlist)


df

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from collections import Counter
from konlpy.tag import Okt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori


reword = []
for i in df['description']:
    okt = Okt()
    pos_text = okt.pos(i)          

    #print(pos_text)

    Noun_word = []                              # 두 글자 이상인 명사만 추출
    for word, pos in pos_text:
        if pos == 'Noun' and len(word) >= 2:
            Noun_word.append(word)

    reword.append(Noun_word)
    
print(reword)
```

```python
te = TransactionEncoder()
te_ary = te.fit(reword).transform(reword)
df_apriori = pd.DataFrame(te_ary, columns=te.columns_)
df_apriori




frequent_itemsets = apriori(df_apriori, min_support=0.64, use_colnames=True)     ######## 지지도 변경
frequent_itemsets





from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=1)
rules
```

## 워드 클라우드

```python
wcStr = ''
for str in Noun_word:
    wcStr = wcStr+str+'\n'

stop_words = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()   # RANKS NL에 제공해주는 한국어 불용어 사전
# 참고 자료 : https://www.ranks.nl/stopwords/korean
stop_words.extend(['위해', '통해', '관련', '지난', '대한', '대해', '이번', '최근', '때문', '이씨', '정연주','홍현기','최자',
                   '관계자','일러스트','라며','폐쇄회','이상','뿌리','제작','다른'])

res1 = re.findall('[가-힣]+', wcStr)                  # 한글을 제외한 모든 글자 제거
res = [w for w in res1 if w not in stop_words]       # stop_words에 있는 글자를 젲외한 나머지 글자 모으기

wcStr_cnt = Counter(res)
wcStr_tuple = wcStr_cnt.most_common()




############################## 워드 클라우드 생성 #####################################

wc = WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",
             background_color="white", width=400, height=400,
             max_words=100, max_font_size=150)

#### 워드 클라우드에 텍스트를 입력해 출력
wcFre = wc.generate_from_frequencies(wcStr_cnt)
plt.figure(figsize=(12, 12))
plt.imshow(wcFre, interpolation='bilinear')
plt.axis('off')
plt.show()
```
